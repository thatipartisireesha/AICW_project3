```python
X=data['Average_Salary']
Y=data['Years_Experience']
plt.scatter(x=X,y=Y,marker='*',c='red',alpha=0.8,cmap='plasma',edgecolors='yellow',linewidths=10)
plt.title("Average_Salary v/s Years_Experience")
plt.xlabel("Average_Salary")
plt.ylabel("Years_Experience")
plt.grid()
plt.show()
plt.scatter(x=X,y=Y,marker='o',c='green',alpha=1.0,cmap='plasma',edgecolors='orange',linewidths=20)
plt.title("Skill_1 v/s Skill_2")
plt.xlabel("Skill_1")
plt.ylabel("Skill_2")
plt.grid()
plt.show()
plt.scatter(x=X,y=data["Average_Salary"],marker='s',c='indigo',alpha=1,cmap='plasma',linewidths=30)
plt.title("Tech_Growth_Factor v/s Automation_Probality_2030")
plt.xlabel("Tech_Growth_Factor")
plt.ylabel("Automation_Probability_Factor")
plt.grid()
plt.show()
```

```python
plt.Figure(figsize=(4,8))
X=data['Skill_3']
Y=data['Skill_4']
plt.scatter(x=X,y=Y,marker='*',c='green',alpha=1.0,cmap='plasma',edgecolors='pink',linewidths=10)
plt.title("Skill_3 v/s Skill_4")
plt.xlabel("Skill_3")
plt.ylabel("Skill_4")
plt.grid()
plt.show()
plt.scatter(x=X,y=Y,marker='o',c='green',alpha=1.0,cmap='plasma',edgecolors='orange',linewidths=20)
plt.title("Skill_5 v/s Skill_6")
plt.xlabel("Skill_5")
plt.ylabel("Skill_6")
plt.grid()
plt.show()
plt.scatter(x=X,y=data["Skill_3"],marker='s',c='indigo',alpha=1,cmap='plasma',linewidths=30)
plt.title("Skill_7 v/s Skill_8")
plt.xlabel("Skill_7")
plt.ylabel("Skill_8")
plt.grid()
plt.show()
```
